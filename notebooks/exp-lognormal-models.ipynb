**Функции для вычисления мощности и ошибки первого рода**

```python
from sklearn.metrics import confusion_matrix

def power_score(y_test, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    power = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    return power

def type1_err_score(y_test, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    type1_error = fp / (fp + tn) if (fp + tn) > 0 else 0.0
    return type1_error
```

#### Запустим логистическую регрессию с ElasticNet

```python
from sklearn.model_selection import GridSearchCV
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(
        penalty='elasticnet',
        solver='saga',
        max_iter=5000,
        random_state=RANDOM_STATE
    ))
])

param_grid = {
    'classifier__C':      [1e-4, 1e-3, 1e-2, 1e-1],
    'classifier__l1_ratio': [0.8, 0.9, 0.95, 1.0]
}
```

```python
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=1
)
grid.fit(X_train, y_train)
```

```python
best_C  = grid.best_params_['classifier__C']
best_l1 = grid.best_params_['classifier__l1_ratio']
print("Лучшие параметры (по F1):", grid.best_params_)
print("Лучший F1 (CV):", grid.best_score_)
```

```python
best_model = grid.best_estimator_
y_pred  = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]
```

```python
print(f'Accuracy:    {accuracy_score(y_test, y_pred):.4f}')
print(f'Precision:   {precision_score(y_test, y_pred):.4f}')
print(f'Recall:      {recall_score(y_test, y_pred):.4f}')
print(f'F1-score:    {f1_score(y_test, y_pred):.4f}')
print(f'ROC AUC:     {roc_auc_score(y_test, y_proba):.4f}')
print(f'Power:       {power_score(y_test, y_pred):.4f}')
print(f'Type 1 Error {type1_err_score(y_test, y_pred):.4f}')
```

```python
coef = best_model.named_steps['classifier'].coef_[0]
mask = coef != 0
selected_features = X_train.columns[mask]

print(f'Отобрано признаков: {mask.sum()} из {X_train.shape[1]}')
print('Список важных признаков:', list(selected_features))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform

param_dist = {
    'svc__C':      loguniform(1e-3, 1e3),
    'svc__kernel': ['linear','rbf'],
    'svc__gamma':  loguniform(1e-4, 1e1)
}

rnd = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=50,
    scoring='f1',             # <-- оптимизируем по F1
    cv=cv,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=2
)

rnd.fit(X_train, y_train)

print("Best params:", rnd.best_params_)
print("Best CV F1:",   rnd.best_score_)
```
